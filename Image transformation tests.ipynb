# Image transformation tests

## A sample image

```python
from matplotlib import pyplot as plt
import cv2
```

```python
sample_image = cv2.imread('data/sample.jpg')
sample_rgb = cv2.cvtColor(sample_image, cv2.COLOR_BGR2RGB)

plt.imshow(sample_rgb)
plt.title('Sample')
plt.show()
```

# A normalized image

```python
from helper import preprocess

normalized = preprocess(sample_image)

plt.imshow(normalized)
plt.title('Normalized')
plt.show()
```

## A flipped image

```python
from helper import flip

flipped_image, angle = flip(normalized, 10, 1)

plt.imshow(flipped_image)
plt.title('Flipped')
plt.show()
```

## A shifted image

```python
from helper import shift

shifted_image, angle = shift(normalized, 10, 100, 10, 0.5)

plt.imshow(shifted_image)
plt.title('Shifted')
plt.show()
```

## A brightnes modified image

```python
from helper import adjust_brightness

brightness_modified = adjust_brightness(normalized, 0.5)

plt.imshow(brightness_modified)
plt.title('Brightness modified')
plt.show()
```

## A shadowed image

```python
from helper import add_shadow

shadowed = add_shadow(normalized)

plt.imshow(shadowed)
plt.title('Shadowed')
plt.show()
```
